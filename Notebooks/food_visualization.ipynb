```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Data/needed_food_data.csv')
```

```python
data.head()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(data, x=data['market'], color='teal')
plt.xticks(rotation=45, ha='right')
plt.title('Total commodities (x100KG) per market')
plt.show()
```

```python
commodity_counts = data['commodity'].value_counts()
explode = (0.05, 0.05, 0.05) 
plt.figure(figsize=(13,13))
plt.pie(commodity_counts, explode = explode, labels=commodity_counts.index, autopct='%1.1f%%', startangle=60, wedgeprops = {"edgecolor" : "black", 'linewidth': 2, 'antialiased': True})
plt.legend()
plt.title('Commodity Distribution')
plt.show()
```

```python
grouping = data.groupby('market')['commodity'].value_counts().unstack()

plt.figure(figsize=(20, 6))
grouping.plot(kind='bar', stacked=True)
plt.xlabel('Market')
plt.ylabel('Count')
plt.title('Commodity Counts by Market')
plt.legend(title='Commodity', loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
year_commodity_avg = data.groupby(['year', 'commodity'])['price'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=year_commodity_avg, x='year', y='price', hue='commodity')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.title('Average Price of Commodities Over the Years')
plt.xticks(rotation=45)
plt.legend(title='Commodity', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.lineplot(data=data, x='year', y='price', hue='commodity')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Price Trends Over Time')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Commodity', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
month_commodity_avg = data.groupby(['month', 'commodity'])['price'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=month_commodity_avg, x='month', y='price', hue='commodity')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.title('Average Price of Commodities Over the Months')
plt.xticks(rotation=45)
plt.legend(title='Commodity', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x="commodity", y="price")
plt.title("Price Distribution by Commodity")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Commodity")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

max_prices = data.groupby("year")["price"].max()
min_prices = data.groupby("year")["price"].min()

plt.stem(max_prices.index, max_prices, basefmt=" ", linefmt="-g", markerfmt="go", label="Max Price")
plt.stem(min_prices.index, min_prices, basefmt=" ", linefmt="-r", markerfmt="ro", label="Min Price")
plt.title("Highest and Lowest Prices by Year")
plt.xlabel("Year")
plt.ylabel("Price")
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
